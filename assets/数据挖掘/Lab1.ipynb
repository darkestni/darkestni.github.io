## Exercise 1: 计算奇数和偶数的和

计算从 1 到 100 的所有奇数和偶数的和

```python

odd_sum = 0
even_sum = 0

for i in range(1, 101):
    if i % 2 == 0:
        even_sum += i
    else:
        odd_sum += i

print(f"1 到 100 的偶数和为：{even_sum}")
print(f"1 到 100 的奇数和为：{odd_sum}")
```

```python

odd_sum = sum(i for i in range(1, 101) if i % 2 != 0)
even_sum = sum(i for i in range(1, 101) if i % 2 == 0)

print(f"1 到 100 的偶数和为：{even_sum}")
print(f"1 到 100 的奇数和为：{odd_sum}")
```

## Exercise 2: 合并重叠区间

实现一个函数，输入一组区间，合并所有重叠的区间并返回结果。

```python
def merge_intervals(intervals):
    
    if not intervals:
        return []
    
    
    intervals.sort(key=lambda x: x[0])
    
    merged = [intervals[0]]  # 初始化结果列表，包含第一个区间
    
    for current in intervals[1:]:
        last = merged[-1]  # 获取结果列表中的最后一个区间
        
        # 如果当前区间的起始位置小于等于最后一个区间的结束位置，说明有重叠
        if current[0] <= last[1]:
            # 合并区间，更新结束位置为两者中的较大值
            last[1] = max(last[1], current[1])
        else:
            # 没有重叠，直接将当前区间加入结果列表
            merged.append(current)
    
    return merged

# 测试示例 1
intervals1 = [[1,3],[2,6],[8,10],[15,18]]
result1 = merge_intervals(intervals1)
print(f"输入：{intervals1}")
print(f"输出：{result1}")
print(f"期望：[[1, 6], [8, 10], [15, 18]]")
print()

# 测试示例 2
intervals2 = [[1,4],[4,5]]
result2 = merge_intervals(intervals2)
print(f"输入：{intervals2}")
print(f"输出：{result2}")
print(f"期望：[[1, 5]]")
```

## Exercise 3: 文件的读写操作

### 3.1 读写 TXT 文件

```python
# 写入 TXT 文件
with open('sample.txt', 'w', encoding='utf-8') as f:
    f.write('这是第一行内容\n')
    f.write('这是第二行内容\n')
    f.write('这是第三行内容\n')
    f.write('Hello, Data Mining!\n')

print('TXT 文件写入成功！')

# 读取 TXT 文件
print('\n--- 读取 TXT 文件内容 ---')
with open('sample.txt', 'r', encoding='utf-8') as f:
    content = f.read()
    print(content)

# 逐行读取
print('--- 逐行读取 ---')
with open('sample.txt', 'r', encoding='utf-8') as f:
    for line in f:
        print(line.strip())  # strip() 去除行末换行符
```

### 3.2 读写 CSV 文件

```python
import csv

# 写入 CSV 文件
with open('sample.csv', 'w', encoding='utf-8', newline='') as f:
    writer = csv.writer(f)
    
    # 写入表头
    writer.writerow(['姓名', '年龄', '城市'])
    
    # 写入数据行
    writer.writerow(['张三', 20, '北京'])
    writer.writerow(['李四', 25, '上海'])
    writer.writerow(['王五', 22, '广州'])

print('CSV 文件写入成功！')

# 读取 CSV 文件
print('\n--- 读取 CSV 文件内容 ---')
with open('sample.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
# 使用 DictWriter 和 DictReader（更直观的方式处理 CSV）
import csv

# 使用字典写入 CSV
with open('sample_dict.csv', 'w', encoding='utf-8', newline='') as f:
    fieldnames = ['姓名', '年龄', '城市']
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    
    writer.writeheader()
    writer.writerow({'姓名': '张三', '年龄': 20, '城市': '北京'})
    writer.writerow({'姓名': '李四', '年龄': 25, '城市': '上海'})
    writer.writerow({'姓名': '王五', '年龄': 22, '城市': '广州'})

print('CSV 文件（字典方式）写入成功！')

# 使用字典读取 CSV
print('\n--- 使用字典方式读取 CSV ---')
with open('sample_dict.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(dict(row))
```
